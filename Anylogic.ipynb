```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np

# Load data from Excel
file_path = '/content/Regression Inflow Ridership for variables.xlsx'  # Change this to your actual file path
sheet_name = 'Sheet1'  # Change this to your actual sheet name

df = pd.read_excel(file_path, sheet_name=sheet_name)

# Define the dependent and independent variables
dependent_var = df.columns[0]
independent_vars = df.columns[1:]

# Perform linear regression for each independent variable
for ind_var in independent_vars:
    X = df[ind_var]
    y = df[dependent_var]

    # Add a constant term for the intercept
    X = sm.add_constant(X)

    # Fit the regression model
    model = sm.OLS(y, X).fit()
    intercept, slope = model.params

    # Scatter plot
    plt.figure(figsize=(8, 6))
    plt.scatter(df[ind_var], df[dependent_var], color='blue', label=f'{ind_var} vs {dependent_var}')

    # Regression line
    line_color = 'green' if slope > 0 else 'red'
    plt.plot(df[ind_var], intercept + slope * df[ind_var], color=line_color, label=f'Regression Line: y={intercept:.2f} + {slope:.2f}x')

    # Title and labels
    plt.title(f'{ind_var} vs {dependent_var}')
    plt.xlabel(ind_var)
    plt.ylabel(dependent_var)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt

# Assuming 'data.xlsx' contains your data
file_path = '/content/Ridership data.xlsx'

# Load data from Excel
df = pd.read_excel(file_path)

# Assuming the first column is 'Year', second column is 'VariableToPredict'
# and remaining columns are input variables
X = df.iloc[:, 2:].values  # Input variables
y = df.iloc[:, 1].values   # VariableToPredict

# Splitting data into training (5 rows) and testing (4 rows)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=4, random_state=42)

# Standardize the data
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)

# Initialize models
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'SVR': SVR(),
    'Random Forest': RandomForestRegressor(random_state=42)
}

# Train and evaluate models
errors = {}
predictions = {}
for name, model in models.items():
    model.fit(X_train_std, y_train)
    y_pred = model.predict(X_test_std)
    predictions[name] = y_pred
    error = mean_absolute_error(y_test, y_pred)
    errors[name] = error

# Determine the best model
best_model_name = min(errors, key=errors.get)
best_model = models[best_model_name]

# Print predicted values and errors
print("Predicted values and errors for the test data:")
for name, y_pred in predictions.items():
    print(f"{name} Predictions: {y_pred}")
    print(f"{name} Error: {errors[name]}")

# Plotting predicted vs actual values
plt.figure(figsize=(10, 6))

# Plot actual values (dashed lines)
plt.plot(df['Year'], df['Total Ridership Trips'], 'b--', label='Actual Values')

# Plot predicted values (thick lines) for all models
for name, model in models.items():
    model.fit(X_train_std, y_train)
    y_pred = model.predict(scaler.transform(X))
    plt.plot(df['Year'], y_pred, label=f'{name} Predictions', linewidth=2 if name == best_model_name else 1)

plt.title('Comparison of Predicted vs Actual Values')
plt.xlabel('Year')
plt.ylabel('Total Ridership Trips')
plt.legend()
plt.grid(True)
plt.show()

# Separate plot for the best algorithm
plt.figure(figsize=(10, 6))
best_model.fit(X_train_std, y_train)
best_y_pred = best_model.predict(scaler.transform(X))
plt.plot(df['Year'], df['Total Ridership Trips'], 'b--', label='Actual Values')
plt.plot(df['Year'], best_y_pred, 'r-', linewidth=2, label=f'{best_model_name} Predictions')
plt.title(f'Best Model: {best_model_name}')
plt.xlabel('Year')
plt.ylabel('Total Ridership Trips')
plt.legend()
plt.grid(True)
plt.show()

# Print the best model
print(f'Best Model: {best_model_name}')
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Load the data from an Excel file
df = pd.read_excel('/content/Regression Inflow Ridership for variables.xlsx')

# Define dependent variable (first column)
y = df.iloc[:, 0].values

# Define independent variables (rest of the columns)
X = df.iloc[:, 1:].values

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize the Ridge regression model with a regularization parameter alpha
ridge_model = Ridge(alpha=1.0)

# Fit the model to the training data
ridge_model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = ridge_model.predict(X_test)

# Calculate and print the mean squared error and R-squared value
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Print the intercept (constant term)
print(f'Intercept: {ridge_model.intercept_}')

# Print the coefficients for each independent variable
coefficients = ridge_model.coef_
for idx, col_name in enumerate(df.columns[1:]):
    print(f'Coefficient for {col_name}: {coefficients[idx]}')
```
